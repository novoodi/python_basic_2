```python
# 1. 필요한 라이브러리 임포트
import pandas as pd
```

```python
# 데이터셋 로드
# dat는 R문법으로 만들어진 함수 
# 읽기 위해서는 read_tacble를 사용해서 만든다.
# : :로 구분해준다, 
# header가 없다.
# R이라는 업어가 사용 된걸 읽기 위해서 engine을 사용한다.
unames = ["user_id", "gender", "age", "occupation", "zip"]
users = pd.read_table("users.dat", sep="::",
                      header=None, names=unames, engine="python")

rnames = ["user_id", "movie_id", "rating", "timestamp"]
ratings = pd.read_table("ratings.dat", sep="::",
                        header=None, names=rnames, engine="python")

mnames = ["movie_id", "title", "genres"]
movies = pd.read_table("movies.dat", sep="::",
                       header=None, names=mnames, engine="python")
```

```python
# 데이터셋 탐색
print(users.head(5))
print(ratings.head())
print(movies.head())
```

## [데이터 분석 문제]

[문제 1]

### # 성별에 따른 250건 이상의 평균 평점 정보가 있는 영화 추출하기

```python
# (1) ratings 테이블과 users 테이블을 병합하기
# (2)그 결과를 movies 테이블과 병합하기 : 1000209 rows × 10 columns
# 공통된 결과를 병합한다. 
# 100만개의 정보/ 10개의 피처
data = pd.merge(pd.merge(ratings, users), movies)
data.head(5)
```

## pandas.pivot_table(data=, values=, index=, colums=, aggfunc=, fill_value=) 간단 개꿀 둘가 기말에 나온다 공부하셈
## groupby("title")['rating'].mean() 기말고사 제출됨 공부 할때 참고 할 것
* 주어진 데이터프레임을 특정 열을 기준으로 그룹화하기
* 다양한 집계 함수를 사용하여 요약 통계를 계산
* 주요 파라미터
  * data : 피벗 테이블을 생성할 데이터프레임
  * values : 집계할 데이터 열
  * index: 피벗 테이블의 행으로 사용할 열
  * columns: 피벗 테이블의 열로 사용할 열
  * aggfunc: 집계함수 목록
  * fill_value: 결측값을 대체할 값

```python
# (1) 성별에 따른 각 영화의 평균 평점을 집계하기
#values 컬럼 이름, 성별에 따른 이기 때문에 gender
mean_ratings = pd.pivot_table(data = data, values = "rating", index="title",
                                columns="gender", aggfunc="mean")
mean_ratings.head(5)
```

```python
# (2) 영화 제목별 건수 집계하기
# 데이터를 영화 제목으로 그룹화하여 size() 함수를 사용
# 동일한 영화 제목을 가지는 행들을 계산하여 반환
# 몇명의 사람들이 봤는지를 찾는 groupby의 메서드
ratings_by_title = data.groupby("title").size()
ratings_by_title.head()
```

```python
# (4) 250건 이상의 평점 정보가 있는 영화 제목 추출
# index의 값이 250 이상인 값을 추출
active_titles = ratings_by_title.index[ratings_by_title >= 250]
active_titles
```

```python
# (5) 성별에 따른 250건 이상의 평점 정보가 있는 영화에 대한 색인 추출하기
# 위에서 평균을 낸 값중에서 250점 이상의 값을 가지고 있는 값을 다시 추출
mean_ratings = mean_ratings.loc[active_titles]
mean_ratings
```

## 문제 2

### # 장르와 연령별로 그룹하여 평균 평점을 구하기

```python
# 여러 장르에 속하는 영화일 경우 장르를 개별적으로 분리하기
# movies["genre"] = movies.gebres.str.split("|") => drop()
movies["genre"] = movies.pop("genres").str.split("|")
movies.head()
```

```python
# (2) 복수 장르에 속하는 영화을 각 장르별로 나누기
movies_exploded = movies.explode("genre")
movies_exploded[:10]
```

```python
# (3) 세 테이블을 모두 병합하여 그룹화하기
ratings_with_genre = pd.merge(pd.merge(movies_exploded, ratings), users)
ratings_with_genre.head(10)
```

```python
ratings_with_genre.shape
```

```python
# (4) 장르와 연령별로 그룹하여 평균 평점을 구하기
# ratings_with_genre.iloc[0]
genre_ratings = (ratings_with_genre.groupby(["genre", "age"])
                 ["rating"].mean().unstack("age")
                 )
genre_ratings[:5]
# genre_ratings_1 = pd.pivot_table(data = ratings_with_genre , values = "rating", index="gender",
#                                 columns="age", aggfunc="mean")
# genre_ratings_1[:5]
```

```python
# (4) 제목과 연령별로 그룹하여 평균 평점을 구하기
# ratings_with_genre.iloc[0]
title_ratings = (ratings_with_genre.groupby(["title", "age"])
                 ["rating"].mean().unstack("age")
                 )
title_ratings.dropna(inplace=True)
title_ratings
```

## [데이터사이언스]
* 선형회귀 분석
  * 각 장르별로 연령대와 평점 간의 관계를 분석하는 선형 회귀 모델을 구축하고, 그 성능을 비교
  * 상관 관계가 있는 데이터 셋을 잘 적합한 장르를 찾기(R^2)로 찾기
  * age를 독립 변수로, 각 장르별 평점을 종속 변수로 사용
  * LinearRegression

```python
# 1. 필요한 라이브러리 임포트
import pandas as pd
import numpy as np
## import하는 것도 알아야한다 언제 어떤 상황에서 쓰이는지
from sklearn.linear_model import LinearRegression
#성능 평가를 위한 r2_score
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt

# 2. 데이터셋 : 데이터프레임 저장
# 재구성한 데이터 프레임
# 
df = genre_ratings
df.describe()

# 3. age와 각 장르의 평점을 각각 독립 변수 X와 종속 변수 y 로 설정
X = np.array([1,18,25,35,45,50,56]).reshape(-1,1)
result=[]
# 4. subplots을 이용하여 각 장르별 선형 회귀 모델링, 성능 평가 및 저장, 실제값과 예측값 비교 시각화
fig, axes = plt.subplots(nrows=9, ncols=2, figsize=(15,45))
axes = axes.flatten()
for idx, genre in enumerate(df.index):
    y = df.loc[genre].values # 장르별로 종속 변수를 만들어 준다.
#LinearRegression 모델을 사용해서 회귀 분석
    model = LinearRegression()
    model.fit(X,y)
    y_pred = model.predict(X)
    #MSE와 R^2
    mse = mean_squared_error(y,y_pred)
    r2 = r2_score(y, y_pred)
    ## 데이터 프레임에 선언
    result.append({
        'genre': genre,
        'mse': mse,
        'r2': r2
    })
    axes[idx].scatter(X,y, color='blue')
    axes[idx].plot(X, model.predict(X), color='red')
    axes[idx].set_title(f'{genre} - Actual va Predicted')
    axes[idx].set_xlabel('age')
    axes[idx].set_ylabel('Rting')
plt.tight_layout()
plt.show()
# 5. 모든 장르의 성능 평가 결과를 데이터프레임으로 저장
result_df = pd.DataFrame(result)
print(result_df)

# 6. 5단계에서 얻은 데이터프레임의 MSE와 R^2 스코어를 막대그래프로 시각화
result_df.set_index('genre', inplace=True)
result_df[['mse', 'r2']].plot(kind='bar', subplots=True, layout=(2,1), figsize=(15,10), title = "MODLE Performanse by Genre")
# 상관 관계가 가장 놓픙 장르 찾기
best_genre = result_df['r2'].idxmax()
print(f'연령과 영화 평점 간의 상관 관계를 가장 작 적합하고 있는 장르: {best_genre}')
```
